```python
# time categories - Early am, late am, early pm, late pm.

def time_cat(time_flt):
    # time categories - Early am, late am, early pm, late pm.
    
    if (time_flt <= 600):
        return 'early_am'
    elif (time_flt <= 1200):
        return 'late_am'
    elif (time_flt < 1800):
        return 'early_pm'
    else:
        return 'late_pm'
    
times = pd.get_dummies(model_db[['DepTime']].applymap(lambda tm: time_cat(tm)))

pd.concat([model_db, times], axis=1)
```

```python
def was_flight_delayed(time_flt):
    if time_flt > 0.0:
        return True
    else:
        return False
    
#pd.get_dummies()
was_delayed = db[['DepDelay']].applymap(lambda tm: was_flight_delayed(tm))
was_delayed = was_delayed.rename(columns={'DepDelay': 'Was_Delayed'})

db = pd.concat([db, was_delayed], axis=1)
db.drop('DepDelay', axis=1)
```
